```python
# imports
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
data = pd.read_csv('Oil and Gas 1932-2014.csv')
data.head()
data.info()
```

```python
#cty name unique values
len(data['cty_name'].unique())

# map of the world with the countries
fig = px.choropleth(data, locations="cty_name", locationmode="country names", color="oil_value_2014", hover_name="cty_name", projection="natural earth", title="Oil and Gas Production 1932-2014")
fig.show()
```
